```python
'''CinemaPass is a program that allows users to book tickets for movies.
The program is stored in the dist folder, which contains all the necessary files 
for the program to run.'''

from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from tkinter.messagebox import showinfo
from PIL import Image,ImageTk
from pathlib import Path
import json
from datetime import date
import copy

# bill function for get booked ticktes in tickets folder in json format
def bill(nw7,session_info,name,ticket,show,phone,dir_movie_info):
    
    #temp_session_info is used to format the dictionary without affecting session_info
    temp_session_info = session_info.copy()
    
    #this block deletes all seats with empty strings in it so final ticket is minimalistic 
    for key1 in temp_session_info["seat"]:
        for key2 in list(temp_session_info["seat"][key1]["matrix"]):
            if temp_session_info["seat"][key1]["matrix"][key2] == "":
                del session_info["seat"][key1]["matrix"][key2]    
                
    #this line creates ticket to be made to printed out 
    ticket.update(session_info)
    
    #makes a Pathlib object that points towards tickets directory which is used to store tickets
    dir_ticket = Path(dir_movie_info).parent.parent.parent / 'Tickets'
    
    
    #this updates the show dictionary and makes it ready to be written off in the 1000.json to stop data leakage
    for seat_type, seat_data in ticket['seat'].items():
        if seat_type in show['seat']:
            show['seat'][seat_type]['matrix'].update(seat_data['matrix'])
    
    
    #dir_movie_info is 1000.json file dir in string
    #dir_movie_info holds path to timing.json which is updated here to reflect the final data 
    with open(dir_movie_info, 'w+') as f:
        json.dump(show, f)
        
    # dir_ticket is Path object that point towards a virtual file phone.json
    #please note the file phone.json is still not created and is a ghost directory 
    dir_ticket = dir_ticket / (phone+'.json')
    
    
    # Check if the file already exists and if it does append numbers at the end
    if dir_ticket.exists():
        i = 1
        #iterates over numbers untill a suitable number is found to be appended
        while (new_file_path := Path(f"Tickets/{phone}-{i}.json")).exists():
            i += 1
        #updates the path that holds phone.json ghost directory 
        dir_ticket = new_file_path
        
    # Create the ticket.json file from ticket dictionary
    with dir_ticket.open("w") as f:
        json.dump(ticket, f)
# seat function for display seats and booked seat
def seat(nw6,film,screen,time,name,phone,dir_movie_info): 
    #dir_movie_info stores timing.json directory 
    #clear screen
    for widget in nw6.winfo_children():
        widget.destroy()
    
    nw6.title(time)
        
    
    #show is dictionary loaded from timing.json
    with open(film["screen"][screen]['shows'][time] ,'r+')as f:
        show = json.load(f)
        
    #checks if movie-info.json and timing.json are related with the relation mentioned in both of the files
    if(film["name"]==show["film"]):
        
        #f is frame that groups seats to be displayed
        f = Frame(nw6,bg="black")
        
        #session info is a dictionary that will be used to hold the seat selection info selected by user 
        #and to be later printed on ticket.json
        session_info=dict()
        session_info['seat']={}
        #makes a deepcopy of timing.json's seat sub-dictionary to get it's structure that will be cleared later on
        session_info['seat']= copy.deepcopy(show['seat'])        
        
        #ticket dictionary will be used to finalize and print it to ticket.json
        ticket={}
        #ticket dictionary is updated with name, phone number, screen data, and film name 
        ticket.update({"name":name,'phone':phone,'screen':screen,'film':film['name']})

#         iterates over seats to clear them 
#         so that we are just left with the dictionary structure of timing.json's seat sub-dictionary
        for i in show['seat'].keys():
            for j in show['seat'][i]['matrix'].keys():
                session_info['seat'][i]['matrix'][j]=''
    
    
#       displays seats with types grouped together by label in i loop    
#         here i iterates over types of seats in timing.json
        for i in show["seat"].keys():
            Label(f,text=i+"="+str(show["seat"][i]["price"])+"\n", font=('Helvetica',20)).pack(pady=100)
#             here j iterates over all the seats in timing.json
            for j in show["seat"][i]["matrix"].keys():
                
#                 switches to disabled state for seat button if it is already booked 
#                 here empty string means the seat is available
                seat_state = DISABLED if show['seat'][i]['matrix'][j]!='' else NORMAL
                
                #this function is used to change the seat to disabled when clicked
                def button_command(button=None):
                    button.configure(state=DISABLED)
                
                #defines button with state and shows number of given seat
                button = Button(f, text=j, bg='gold', state=seat_state, width='5', height='5')
                #configures on click function which puts selected seat data in session info variable 
#                 and with lambda function we also call button_command function that disables the given seat
                button.configure(command=lambda button=button,seat_type=i,seat_no=j:(session_info['seat'][seat_type]['matrix'].update({seat_no: name}), button_command(button=button)))
                button.pack(side=LEFT, padx=10)
        
        #submit button which calls the next function bill() with required variables 
        Button(nw6, text="submit", command=lambda session_info=session_info, show=show, ticket=ticket: bill(nw6, session_info, name, ticket, show,phone,dir_movie_info)).pack(side=BOTTOM, pady=(0,70))
        f.pack()
    else:
        #if the relation mentioned in both files does not match then an error shows
        print("movie name in movie-info.json and movie name in screens/.../timing.json does not match")
        
# show function displays posters of movies that can be clicked on to select them
def show(nw4,film,name,phone):
    
    #clear screen
    for widget in nw4.winfo_children():
        widget.destroy()
        
    #here film is Pathlib object that points towars your selected film
    nw4.title(film)
    #here film is overridden as a python dictionary from accessing it's Movie-info.json file
    with open(film / "Movie-info.json" ,'r+')as f:
        film = json.load(f)
        
#     here i iterates over available screens in Movie-info.json
    for i in film['screen']:
        #groups different timing in different screens 
        Label(nw4,text= "Screen : " + i , font=('Helvetica',20)).pack(pady=100)
        #here j iterates over shows which is nested with i variable which holds screen data for that film
        for j in film['screen'][i]['shows']:
            #creates Tkinter button object for diffent timings and intiates the next step of the program
            #film is movie-info.json
            #i is screen
            #j is timing
            #and the last variable holds path towards timing.json in Screens folders subdirectories 
            ttk.Button(nw4, text=j, command=lambda i=i,j=j: seat(nw4,film,i,j,name,phone,film['screen'][i]['shows'][j])).pack(pady=20)
            
            
# nextscreen function displays movie images as button to click image  
def NextScreen(nw1,path,name,phone):
    
#   destroy previous screen widgets  
    for widget in nw1.winfo_children():
        widget.destroy()
        
    #uses path variable to access Films folder that later will be used to show poster and get Movie-info.json 
    films = path / 'Films'

    # Get the names of directories in the Films directory
    dirs = [x for x in films.iterdir() if x.is_dir()]
    
    
    Label(nw1,text="Click on image to book tickets !!", font=('Helvetica',20)).pack(pady=20)
    
    #canvas is used to group movie poster for better display
    canvas = Canvas(nw1, width = 1000, height = 800).pack(expand=YES)
    
    #the following code is used to fetch Posters from Films folder dynamically 
    #and make them ready to be used as Tkinter Objects 
    padding = 20
    for i in [x for x in films.iterdir() if x.is_dir()]:
        photo = ImageTk.PhotoImage(Image.open(i / 'Poster.jpeg').resize((500,750)))
        poster = ttk.Button(canvas,text=i,image= photo)
        poster.image=photo
        poster.place(x = padding , y = 70)
        padding+=photo.width()
        #bind method is used to bind mouse left click(<Button-1>) to poster objects
        #here i is directory of your selected film which is mentioned in show() as film
        poster.bind("<Button-1>", lambda event,i=i: show(nw1,i,name,phone))
# this function validate phone no and username
def validate_phone(nw2,path,phone_number,uname):
    #next line checks the phone_number is digits only and exactly 10 letters long
    if phone_number.get().isdigit() and len(phone_number.get()) == 10:
        #gives alert for invalid name if name = empty string
        if uname.get() != '':
            name = uname.get()
            phone = phone_number.get()
            NextScreen(nw2,path,name,phone)
        else:        
            messagebox.showinfo("Alert", "Not Valid Name", icon="warning")
    else:
        messagebox.showinfo("Alert", "Not Valid Phone Number", icon="warning")
        
        
#     this function is home screen display image,title,icon and inputs like name,phone no.
def home(nw3):
    # Get the current working directory with pathlib Path object 
    path = Path(".")
#     apply title,dimension and full screen state to window object
    nw3.title("CinemaPass : Your one stop for faster tickets")
    nw3.geometry("700x700")
    nw3.state('zoomed')
    
#     set icom
    nw3.iconbitmap(path / 'icon-low.ico')
    
    #layout for home screen 
    #image/photo = CinemaPass logo
    image = Image.open(path / 'icon-high.png')
    resized_image= image
    photo = ImageTk.PhotoImage(resized_image)
    #label that holds logo as an image
    i1 = Label(nw3,image=photo)
    i1.image = photo
    i1.pack(side=LEFT, padx=(200,0))
    
    #credentials is a canvas that holds user details visual elements 
    credentials = Canvas(nw3)
#     Tkinter Entry object for username
    Label(credentials,text="Enter your name", font=('Helvetica',20)).pack(pady=20)
    uname= Entry(credentials,width=20)
    uname.pack()
#     Tkinter Entry object for phone number
    Label(credentials,text="Enter your phone number", font=('Helvetica',20)).pack(pady=20)
    phone_number = Entry(credentials, width=20)
    phone_number.pack()
    
#     button that initiate the next step of program 
    Button(credentials, text="Submit", command=lambda:validate_phone(nw3,path,phone_number,uname)).pack(pady=23)
    credentials.pack(side=RIGHT,padx=(0,300))
    
    
# creates Tkinter winnow object
window = Tk()
# call home screen and pass window as an argument
home(window)
window.mainloop()
```
